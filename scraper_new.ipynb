```python
# Import selenium 
from selenium import webdriver
import pandas as pd 
from bref_stats import *

urls = create_bref_urls()
```

```python
del urls['Manny Ramirez']
urls
```

```python
from tqdm import tqdm 

# Create a new instance of the Firefox driver

all_data = []
# Navigate to the application home page

for player in tqdm(urls): 
    url = urls[player]
    try: 
        driver = webdriver.Firefox()
        driver.get(url)

        element = driver.find_elements(by='xpath', value='//div[@id="div_batting_value"]')

        all_text = [x.text for x in element]

        one_big_str = '\n'.join(all_text)

        lines = one_big_str.split('\n')

        newlines = []
        for line in lines: 
            newlines.append(line.split(' '))

        data = pd.DataFrame(newlines[1:])
        data.columns = newlines[0]
        data['player'] = player

        all_data.append(data)
        
        driver.close()
    except:
        print(url)
        pass
```

```python
test = pd.concat(all_data)
test.to_csv('new_scraper_data.csv')
```
