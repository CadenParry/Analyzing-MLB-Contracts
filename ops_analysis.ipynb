```python
contracts = pd.read_csv('contracts.csv', sep=',')
Batting_csv = pd.read_csv('kaggle/Batting.csv', sep=',')


def get_player_id(player_name):
    player_id_list = []
    name_parts = player_name.split()  # Split the name into parts
    # Handle names with different number of parts
    if len(name_parts) == 2:  # If the name has two parts
        first_name, last_name = name_parts
        suff = ''
    elif len(name_parts) == 3:  # If the name has three parts
        first_name, last_name, suff = name_parts
    else:
        print(f"Unexpected name format: {player_name}")

    if len(last_name) > 5:
        player_id = (last_name[0:5] + first_name[0:2]).lower() + str(0)
        for i in range(1, 5):
            player_hold = player_id + str(i)
            player_id_list.append(player_hold)
    else:
        player_id = (last_name + first_name[0:2]).lower() + str(0)
        for i in range(1, 10):
            player_hold = player_id + str(i)
            player_id_list.append(player_hold)

    return player_id_list

for player in contracts['player']:

    possible_player_ids = get_player_id(player)

    valid_ids = []
    for player_id in possible_player_ids:
        if player_id in batting_id:
            valid_ids.append(player_id)

    if len(valid_ids) == 0:
        print(f"Player {player} not found in batting data")

    if len(valid_ids) == 1:
       # Find the row where the player is Alex Rodriguez
        row = contracts[contracts['player'] == str(player)]

        # Update the player_id field for that row
        contracts.loc[row.index, 'player_id'] = valid_ids[0]  # Replace 'arod01' with the actual player_id

        # Save the DataFrame back to the CSV file
        contracts.to_csv('contracts.csv', sep=',', index=False)
      
    if len(valid_ids) > 1:
        for id in valid_ids:
            #print(player + " " + id)
            continue
```
