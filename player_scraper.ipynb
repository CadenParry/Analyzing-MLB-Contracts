```python
urls = urls[1].values[1:]
```

```python
def get_player_stats(url):
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')

    tables = soup.find_all('table', {'class': 'salaryTable'})
    a = soup.find_all('th', {'class': 'header center'})[6]
    relevant_table = a.parent.parent.parent

    all_rows = relevant_table.find_all('tr', {'class': 'salaryRow'})
    cols = [x.text for x in all_rows[0].find_all('th')]
    vals = []
    for row in all_rows[1:]: 
        vals.append([x.text for x in row.find_all('td')])

    import pandas as pd 
    df = pd.DataFrame()
    df.index = cols
    for i, val in enumerate(vals): 
        df[i] = val

    player_name = urls[0].split('/')[-3]
    df.transpose().to_csv('data/%s.csv' % player_name)
```

```python
# url = 'https://www.spotrac.com/mlb/los-angeles-angels/mike-trout-8553/statistics/'

for url in tqdm(urls): 
    try: 
        get_player_stats(url)
    except Exception as e: 
        print(e)
        print(url)
        continue
```
